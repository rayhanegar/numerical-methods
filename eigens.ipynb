### Metode Power

```python
# constants
RAND_LOW = -16
RAND_HIGH = 16
ITERATION = 100
SIZE = 3

a = np.zeros((SIZE,SIZE))
b = np.zeros(SIZE)

for i in range(SIZE):
    b[i] = r.randint(RAND_LOW, RAND_HIGH)
    # for j in range(SIZE):
    #     a[i,j] = r.randint(RAND_LOW,RAND_HIGH)

# a = np.array([[1,1,0],[1,0,1],[0,1,1]])
a = np.array([[2,-3,0],[2,-5,0],[0,0,3]])

print("known:")
print("a=", a)
print("b=", b)

def mult(a,b):
    n = len(a)
    c = np.zeros(n)
    for i in range(n):
        for j in range(n):
            c[i]+=a[i,j]*b[j]
    # print("a*b=",c)
    return c

def div_with_max(c,d):
    n = len(c)
    e = np.zeros(n)
    for i in range(n):
        e[i]=c[i]/d
    return e

# print(mult(a,b))
# print(max(mult(a,b)))

def power(a,b):
    A = np.copy(a); B = np.copy(b)
    for i in range(ITERATION):
        c = mult(A,B)
        d = max(abs(c))
        # print("d=",d)
        B = div_with_max(c,d)
    return d,B

d,b=power(a,b)
print("\ndominant eigenvalue=", d)
print("dominant eigenvector=", b)
```
