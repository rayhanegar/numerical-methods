```python
import numpy as np
import matplotlib.pyplot as plt  
```

```python
# contoh-contoh soal
test_1 = lambda x: 5*x**2 - 2
test_2 = lambda x: 2*np.exp(np.log(x**2))
test_3 = lambda x: np.sin(x)

# a = batas bawah interval, b = batas atas, n = step
a = 0
b = 2*np.pi
n = 20

# option = forward, backward, central; ketik di bawah bukan di sini
```

```python
def forward_diff(x,y):
    df = [(y[i+1]-y[i])/(x[i+1]-x[i]) for i in range(len(x)-1)]
    X = [x[i] for i in range(len(x)-1)]
    return df,X

def back_diff(x,y):
    df = [(y[i]-y[i-1])/(x[i]-x[i-1]) for i in range(1,len(x))]
    X = [x[i+1] for i in range(len(x)-1)]
    return df,X

def central_diff(x,y):
    df = [(y[i+1]-y[i-1])/(x[i+1]-x[i-1]) for i in range(1,len(x)-1)]
    X = [x[i] for i in range(1,len(x)-1)]
    return df,X
```

```python
def my_der_calc(f,a,b,n,option):
    x = np.linspace(a,b,n)
    y = [f(x[i]) for i in range(len(x))]

    if option == 'forward':
        df,X = forward_diff(x,y)
    elif option == 'backward':
        df,X = back_diff(x,y)
    elif option == 'central':
        df,X = central_diff(x,y)
    else: df,X=[0]
    return df, X
```

```python
def my_num_diff(f,a,b,n,option):
    dy,X = my_der_calc(f,a,b,n,option)
    return dy, X
```

Jangan lupa ganti parameter `f = test_1, 2, 3,...` sesuai dengan soal dan linspace untuk menghindari error.

No. 2 (plotting hanya di no. 2)

```python
[dyf, Xf] = my_num_diff(test_3, a, b, n, 'forward')
[dyb, Xb] = my_num_diff(test_3, a, b, n, 'backward')
[dyc, Xc] = my_num_diff(test_3, a, b, n, 'central')
x_analytic = np.linspace(a,b,n)
plt.figure(figsize = (12, 8))
plt.plot(x_analytic, np.cos(x_analytic), label = 'analytic')
plt.plot(Xf, dyf, label = 'forward')
plt.plot(Xb, dyb, label = 'backward')
plt.plot(Xc, dyc, label = 'central')
plt.legend()
plt.title('Analytic and Numerical Derivatives of Sine')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
